```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics

# seaborn to better visualization
import seaborn as sns
sns.set()

plt.rcParams['figure.figsize'] = [10, 12]
plt.rcParams['figure.dpi'] = 80 # 200 e.g. is really fine, but slower
```

```python
methods = ['SIGMOIDE', 'THAN']
tanh_x = []
sigmoide_x =[]
for m in methods:
    for i in range(1,9):
        filename = f'{m}-{i}.txt'
        with open(filename) as f:
            print(filename)
            values = list(map(float, f.readlines()))
            if m == "THAN":
                tanh_x.append(values)
            else:
                sigmoide_x.append(values)
```

```python
# epochs
y = list(range(1,21))
```

```python
fig, (ax0, ax1) = plt.subplots(2)
fig.tight_layout() # add space between plots

# ReLU
# ax0.plot(x, y,label="sin")
# ax0.plot(x, y,label="sin")
# ax0.plot(x, y,label="sin")
# ax0.plot(x, y,label="sin")
# ax0.set_title("ReLU")
# ax0.legend(loc='best')

# Tanh
for layers, x in enumerate(tanh_x):
    ax0.plot(y, x, label=f"with-{layers+1}-layers")
ax0.set_title("Tanh")
ax0.legend(loc='best')
ax0.set(ylabel="error")

# Sigmoid
for layers, x in enumerate(sigmoide_x):
    ax1.plot(y, x, label=f"with-{layers+1}-layers")
ax1.set_title("Sigmoid")
ax1.legend(loc='best')
ax1.set(xlabel="epochs", ylabel="error")
```
